```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
# make datetime string into datetime object
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
# make string list in column job_skills into panda series (list) object
df["job_skills"] = df["job_skills"].apply(lambda skill_list: ast.literal_eval(skill_list) if pd.notna(skill_list) else None)
```

```python
job_list = ["Data Analyst", "Data Engineer", "Data Scientist"]

df_UK = df[(df["job_title_short"].isin(job_list)) & (df["job_country"] == "United Kingdom")].copy()

df_UK
```

```python
df_UK_exploded = df_UK.explode("job_skills")

df_UK_exploded
```

```python
df_UK_exploded.groupby("job_skills")["job_skills"].agg(["count"]).sort_values(by="count", ascending=False).head(5)

df_UK_exploded.groupby("job_skills")["job_skills"].agg(["count"]).sort_values(by="count", ascending=False)
```
