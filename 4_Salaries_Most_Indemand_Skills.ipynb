Import Libraries such as Pandas, Datasets, Matplotlib.pyplot and Seaborn.

```python
# Importing Libraries
import ast # For string conversion into list object
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick # To add percentage sign to y axis ticks

plt.close('all') # close all open plots

import seaborn as sns

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
# make datetime string into datetime object
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
# in column job_skills, make list from string into panda series (list) object using ast.literal_eval
df["job_skills"] = df["job_skills"].apply(lambda skill_list: ast.literal_eval(skill_list) if pd.notna(skill_list) else None)
```

```python
# In bar plots, FuncFormatter is to format x axis ticks as £K
from matplotlib.ticker import FuncFormatter
def pounds_k(x, pos):
    return f"£{x/1000:.0f}K"
```

# What are the median salaries of top 5 most in-demand skills?

Filter the dataframe to show job postings from the United Kingdom whilst ensuring to create a copy of it also for data integrity. I also drop rows containing NA values as they are not relevant for my analysis.

```python
df_UK = df[df["job_country"] == "United Kingdom"].copy()

df_UK = df_UK.dropna(subset="salary_year_avg")

df_UK
```

As I have made the values in column job_skills from string into list object, they are now 'explodable' and can be exploded into multiple rows. Below they are exploded and assigned to variable df_UK_exploded.

```python
df_UK_exploded = df_UK.explode("job_skills")

df_UK_exploded
```

I aggregate the data using .groupby() function and count the mention of skills in job postings using 'count' as the aggegation method, and calculate the median yearly salary using 'median' as the aggregation method.

```python
df_UK_top_skills = (df_UK_exploded
                    .groupby("job_skills")["salary_year_avg"]
                    .agg(["count", "median"])
                    .sort_values(by="count", ascending=False)
                )

df_UK_top_skills
```

```python
# Salaries of top 5 skills and plot them on bar plot
salaries_of_top_5_skills = df_UK_top_skills["median"].head()

# Plot using Matplotlib
# Values are resorted to ensure highest value is at top
salaries_of_top_5_skills.sort_values().plot(
    kind="barh",
    x="job_skills",
    y="median"
)

plt.gca().xaxis.set_major_formatter(FuncFormatter(pounds_k)) # Format x axis ticks as £K

plt.title("Salaries of Top 5 skills for Data Jobs")
plt.ylabel("Skill")
plt.tight_layout() # Makes spacing nice
```

```python
df_salaries = salaries_of_top_5_skills.sort_values(ascending=False).reset_index()

sns.set_theme(style="whitegrid", font_scale=1.1)
plt.figure(figsize=(9, 5))

sns.barplot(
    data=df_salaries, # reset index to turn into dataframe, otherwise plot fails
    x="median",
    y="job_skills",

    hue="job_skills",
    palette="dark:teal_r"
     
);

plt.gca().xaxis.set_major_formatter(FuncFormatter(pounds_k)) # Format x axis ticks as £K

plt.title("Salaries of Top 5 skills")
plt.ylabel("")
plt.ylabel("Skill")
plt.tight_layout() # Makes spacing nice
```
